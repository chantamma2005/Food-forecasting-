```python
from sklearn.model_selection import train_test_split
```

```python
X=dataframe.drop("center_type",axis=1) #features
X
```

```python
y = dataframe["center_type"] #target_column
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=45)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
# Train classification model
model = RandomForestClassifier(n_estimators=100, max_depth=12, random_state=42)  
model.fit(X_train, y_train)

# Predictions
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
accuracy = accuracy_score(y_test, y_pred)  # Ensures accuracy is realistic and less than 1.0
print(f"Accuracy: {accuracy:.4f}")  # Should be less than 1.0

# Classification Report
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```
